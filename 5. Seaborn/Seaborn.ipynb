# Seaborn

Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics.

- Provides a layer of abstraction hence simpler to use
- better aesthetics
- more graphs included

### Main Components of Seaborn 
Types of Functions 
1. Figure Level
2. Axis Level

Main Classification
- Relational Plot
- Distribution Plot
- Categorical Plot
- Regression Plot
- Matrix Plot
- Multiplots

```python
import seaborn as sns
import numpy as np
import plotly.express as px
```

## 1. Relational Plot
Generally used for 
- see the statistical relation between 2 or more variables.
- Bivariate Analysis

Figure level - relplot()  
Axis level -
1. scatterplot()
2. lineplot()

```python
tips = sns.load_dataset('tips')
tips
```

### sns.scatterplot(data, x, y, hue, style, size)

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='time', size='size')
```

#### sns.relplot(data, x, y, kind, hue, style, size)

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter', hue='sex', style='time', size='size')
```

### sns.lineplot(data, x, y)

```python
gap = px.data.gapminder()
temp = gap[gap['country'] == 'India']
temp
```

```python
sns.lineplot(data=temp, x='year', y='lifeExp')
```

#### sns.relplot(data, x, y, kind, hue, style, size)

```python
sns.relplot(data=temp, x='year', y='lifeExp', kind='line')
```

```python
temp = gap[gap['country'].isin(['India', 'Brazil', 'Germany'])]
temp
```

```python
sns.relplot(data=temp, x='year', y='lifeExp', kind='line', hue='country', style='continent', size='continent')
```

#### Facet Plot
Figure level function, only work with relplot

```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='line', col='sex', row='day')
```

```python
# Col Wrap
sns.relplot(data=gap, x='lifeExp', y='gdpPercap', kind='scatter', col='year', col_wrap=2)
```

## 2. Distribution Plots
Generally used for 
- Univariate Analysis
- used to find out the distribution of data
- shows the range of the data
- central tendency
- used to check whether there are any outliers or the data bimodal?

Figure level - displot()  
Axis level -
1. histplot()
2. kdeplot()
3. rugplot()

### sns.histplot(data, x, bins, hue, element)
element is used to designing

```python
sns.histplot(data=tips, x='total_bill', bins=10, hue='sex', element='step')
```

#### sns.displot(data, x, kind)

```python
sns.displot(data=tips, x='total_bill', kind='hist')
```

```python
# Categorical variables can also be plotted

sns.displot(data=tips, x='day', kind='hist')
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.displot(data=titanic, x='age', kind='hist', element='step', hue='sex')
```

#### Facet Plot

```python
sns.displot(data=tips, x='tip', kind='hist', col='sex', element='step')
```

#### Bivariate Histogram
A bivariate histogram bins the data within rectangles that tile the plot and then shows the count of observations within each rectangle with the fill color

```python
# sns.histplot(data=tips, x='total_bill', y='tip')
sns.displot(data=tips, x='total_bill', y='tip', kind='hist')
```

### sns.kdeplot(data, x, hue, fill)
Rather than using discrete bins, a KDE plot smooths the observations with a Gaussian kernel, producing a continuous density estimate

```python
sns.kdeplot(data=tips, x='total_bill')
```

```python
sns.kdeplot(data=tips, x='total_bill', hue='sex', fill=True)
```

#### sns.displot(data, x, kind)

```python
sns.displot(data=tips, x='total_bill', kind='kde')
```

#### Bivariate Kdeplot
A bivariate kde plot smoothes the (x,y) observations with a 2D Gaussian

```python
sns.kdeplot(data=tips, x='total_bill', y='tip')
```

### sns.rugplot(data, x)
Plot marginal distributions by drawing ticks along the axes.  
This function is intended to complement other plots by showing the location of individual observations in an unobtrusive way

```python
sns.kdeplot(data=tips, x='total_bill')
sns.rugplot(data=tips, x='total_bill')
```

## 3. Matrix Plots

Figure level - displot()  
Axis level -
1. heatmap()
2. clustermap()

### sns.heatmap(data, annot=False, linewidth, cmap)

```python
temp = gap.pivot(index='country', columns='year', values='lifeExp')
sns.heatmap(temp)
```

```python
temp = gap[gap['continent'] == 'Europe'].pivot(index='country', columns='year', values='lifeExp')
import matplotlib.pyplot as plt
plt.figure(figsize=(15,15))
sns.heatmap(temp, annot=True, linewidth=0.5, cmap='viridis')
```

### sns.clustermap(data, annot=False, linewidth, cmap)
Plot a matrix dataset as a hierarchically-clustered heatmap.  
This function requires scipy to be available.

```python
import plotly.express as px
iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:, [0,1,2,3]])
```

## Plotting Pairwise Relationship (PairGrid vs Pairplot)

```python
sns.pairplot(iris, hue='species')
```

```python
# pair grid
g = sns.PairGrid(data=iris, hue='species')
# g.map
g.map(sns.scatterplot)
```

```python
# map_diag -> map_offdiag

g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.boxplot)
g.map_offdiag(sns.kdeplot)
```

```python
# map_diag -> map_upper -> map_lower
g = sns.PairGrid(data=iris, hue='species')
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```

```python
# vars
g = sns.PairGrid(data=iris,hue='species',vars=['sepal_width','petal_width'])
g.map_diag(sns.histplot)
g.map_upper(sns.kdeplot)
g.map_lower(sns.scatterplot)
```
